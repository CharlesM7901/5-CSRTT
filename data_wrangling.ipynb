```python
import pandas as pd

# Open the master_sheet.xlsx file - Jour 3 sheet
master_sheet = pd.read_excel('master_sheet.xlsx', sheet_name='Jour 3')

# Display the first few rows
master_sheet.head()
```

```python
# Remove columns that are entirely empty (all NaN/null values)
columns_to_drop = [col for col in master_sheet.columns if master_sheet[col].isna().all()]
master_sheet = master_sheet.drop(columns=columns_to_drop)

print(f"Dropped {len(columns_to_drop)} empty column(s): {columns_to_drop}")
print(f"\nRemaining columns: {list(master_sheet.columns)}")
print(f"\nCleaned data shape: {master_sheet.shape}")
master_sheet.head()
```

```python
# Open the file_names.xlsx file
file_names = pd.read_excel('file_names.xlsx')

# Extract components from the "Name" column (format: YYYY_MM_DD__hh_mm_ss_animal_number.csv)
file_names[['Year', 'Month', 'Day', 'Hour', 'Minute', 'Second', 'Animal_Number']] = file_names['Name'].str.extract(
    r'(\d{4})_(\d{2})_(\d{2})__(\d{2})_(\d{2})_(\d{2})_(.+?)\.csv'
)

# Convert numeric columns to integers
numeric_cols = ['Year', 'Month', 'Day', 'Hour', 'Minute', 'Second']
for col in numeric_cols:
    file_names[col] = pd.to_numeric(file_names[col], errors='coerce').astype('Int64')

print("Extracted columns:")
print(file_names)
print(f"\nColumns: {list(file_names.columns)}")

# Save the updated dataframe back to the same file
file_names.to_excel('file_names.xlsx', index=False)
print("\nFile saved successfully to file_names.xlsx")
```

```python
# Define the date mapping for each Jour
date_mapping = {
    'Jour 1 (trial run)': (2025, 12, 9),
    'Jour 1': (2025, 12, 10),
    'Jour 2': (2025, 12, 11),
    'Jour 3': (2025, 12, 12),
    'Jour 4': (2025, 12, 14),
    'Jour 5': (2025, 12, 15),
    'Jour 6': (2025, 12, 16),
    'Jour 7': (2025, 12, 17),
    'Jour 8': (2025, 12, 18),
    'Jour 9': (2025, 12, 19),
    'Jour 10': (2025, 12, 20)
}

# Reload file_names to ensure we have the latest data
file_names_fresh = pd.read_excel('file_names.xlsx')
file_names_fresh[['Year', 'Month', 'Day', 'Hour', 'Minute', 'Second', 'Animal_Number']] = file_names_fresh['Name'].str.extract(
    r'(\d{4})_(\d{2})_(\d{2})__(\d{2})_(\d{2})_(\d{2})_(.+?)\.csv'
)
file_names_fresh[['Year', 'Month', 'Day']] = file_names_fresh[['Year', 'Month', 'Day']].astype(int)

# Get the current sheet name and date
current_jour = 'Jour 3'
year, month, day = date_mapping[current_jour]

# Function to match subject ID to filename for specific date
def get_filename_for_subject_date(subject_id, file_df, year, month, day):
    """Extract animal number from subject ID and find matching filename for the date"""
    # Remove letters like V, *, etc. to get just the number
    animal_number = ''.join(c for c in str(subject_id) if c.isdigit())
    
    # Filter file_names for the specific date and animal number
    match = file_df[(file_df['Year'] == year) & 
                    (file_df['Month'] == month) & 
                    (file_df['Day'] == day) & 
                    (file_df['Animal_Number'] == animal_number)]
    
    if len(match) > 0:
        return match.iloc[0]['Name']
    return None

# Get the first subject ID from master_sheet
first_subject = master_sheet.iloc[0, 0]
print(f"First subject: {first_subject}")
print(f"Current Jour: {current_jour}")
print(f"Date: {year}/{month:02d}/{day:02d}")

# Find matching filename
filename = get_filename_for_subject_date(first_subject, file_names_fresh, year, month, day)
print(f"Matching filename for subject {first_subject}: {filename}")

# Now add the new column at the front of master_sheet
master_sheet.insert(0, 'Data_File', '')

# Fill in the filenames for each row
for idx, row in master_sheet.iterrows():
    subject_id = row.iloc[1]  # Second column now (was first before insert)
    filename = get_filename_for_subject_date(subject_id, file_names_fresh, year, month, day)
    master_sheet.at[idx, 'Data_File'] = filename

print("\nFirst few rows with new Data_File column:")
print(master_sheet.head())
```

```python
# Save a copy of the updated master_sheet with the new name
master_sheet.to_excel('master_sheet_names.xlsx', sheet_name='Jour 3', index=False)
print("Master sheet saved as 'master_sheet_names.xlsx'")
```
